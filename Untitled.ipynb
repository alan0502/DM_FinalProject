```python
# Import the data
import pandas as pd
train_df = pd.read_csv("child-mind-institute-problematic-internet-use/train.csv")
test_df = pd.read_csv("child-mind-institute-problematic-internet-use/test.csv")
```

```python
train_df
```

```python
train_df.drop(['Basic_Demos-Enroll_Season', 'CGAS-Season', 'Physical-Season', 'Fitness_Endurance-Season', 'FGC-Season', 'BIA-Season', 'PAQ_A-Season', 'PAQ_C-Season', 'PCIAT-Season', 'SDS-Season', 'PreInt_EduHx-Season'], axis=1, inplace=True)
test_df.drop(['Basic_Demos-Enroll_Season', 'CGAS-Season', 'Physical-Season', 'Fitness_Endurance-Season', 'FGC-Season', 'BIA-Season', 'PAQ_A-Season', 'PAQ_C-Season', 'SDS-Season', 'PreInt_EduHx-Season'], axis=1, inplace=True)
```

```python
train_df
```

```python
train_df
```

```python
# 使用每列的眾數填補 NaN
for column in train_df.columns:
    train_df[column].fillna(train_df[column].mode()[0], inplace=True)
```

```python
train_df
```

```python
# 使用每列的眾數填補 NaN
for column in test_df.columns:
    test_df[column].fillna(test_df[column].mode()[0], inplace=True)
```

```python
X_train = train_df.drop(columns=['sii'])  # 假設 'category' 是標籤欄位
y_train = train_df['sii']

X_test = test_df
y_test = pd.read_csv("child-mind-institute-problematic-internet-use/sample_submission.csv")
```

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

# 確保訓練和測試數據的欄位一致
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score

# 初始化模型
model = RandomForestClassifier(random_state=42)

# 訓練模型
model.fit(X_train, y_train)

# 預測測試集
y_pred = model.predict(X_test)

# 計算準確率
accuracy = accuracy_score(y_test['sii'], y_pred)
print(f"Accuracy: {accuracy}")

# 輸出分類報告
report = classification_report(y_test['sii'], y_pred)
print("Classification Report:\n", report)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test['sii'], y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42)
# 訓練模型
model.fit(X_train, y_train)

# 預測測試集
y_pred = model.predict(X_test)

# 計算準確率
accuracy = accuracy_score(y_test['sii'], y_pred)
print(f"Accuracy: {accuracy}")

# 輸出分類報告
report = classification_report(y_test['sii'], y_pred)
print("Classification Report:\n", report)
```
